```python
tdf = pd.read_csv('team_data_clean_v2.csv')
pdf = pd.read_csv('player_data_clean_v2.csv')
```

```python
#line up home and away teams so there is 1 row for each game
homedf = tdf.loc[tdf['MATCHUP'].str.contains('vs')]

home_dict = dict(zip(homedf.columns, [x + '_home' for x in homedf.columns]))

homedf = homedf.rename(columns=home_dict).drop(columns='MATCHUP_home')
homedf
```

```python
#line up home and away teams so there is 1 row for each game
awaydf = tdf.loc[~(tdf['MATCHUP'].str.contains('vs'))]

away_dict = dict(zip(awaydf.columns, [x + '_away' for x in awaydf.columns]))

awaydf['TEAM_home'] = awaydf['MATCHUP'].apply(lambda x: x[x.index('@_') + 2:])

awaydf = awaydf.rename(columns=away_dict)#.drop(columns='MATCHUP_away')
awaydf
```

```python
#awaydf.rename(columns={'GAME_DATE_away': 'GAME_DATE_home'}, inplace=True)
#awaydf.columns, homedf.columns
homedf.drop_duplicates(inplace=True)
awaydf.drop_duplicates(inplace=True)

merged_df = homedf.merge(awaydf, on=['TEAM_home', 'GAME_DATE_home'])
merged_df
```

```python
merged_df.to_csv('team_data_clean_v3.csv', index=False)
```
